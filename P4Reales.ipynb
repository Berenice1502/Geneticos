```python
#Elaboro: Berenice Francisco Jiménez - ICO23
import random
import math

def funcionObjetivo(x):
    fx=0
    V=4189.829101
    for i in range(len(x)):
        fx+=-x[i]*math.sin(math.sqrt(abs(x[i])))
    return round((10*V)+fx,4)

def Poblacion(tam):
    poblacion=[]
    noGen=11
    valorX=[]
    valorY=[]
    aptitud=[]
    for i in range(tam):
        x=[]
        i=0
        while i<noGen:
            n=random.uniform(-500,500)
            if n in x==True:
                continue
            x.append(n)
            i+=1
        valorX.append(x)
        valorY.append(funcionObjetivo(x))
        aptitud.append(-1)
    poblacion.append(valorX)
    poblacion.append(valorY)
    poblacion.append(aptitud)
    return poblacion

def Evaluacion(x1,tam):
    fx,aptitud,nuevaPoblacion=[],[],[]
    for i in range(tam):
        fx.append(funcionObjetivo(x1[i]))
        aptitud.append(-1)
    nuevaPoblacion.append(x1)
    nuevaPoblacion.append(fx)
    nuevaPoblacion.append(aptitud)
    BubbleSort(nuevaPoblacion,tam)
    calcularAptitud(nuevaPoblacion,tam)
    return nuevaPoblacion
        
    
def calcularAptitud(poblacion,tam):
    sumatoria=0
    for i in range(tam):
        sumatoria+=poblacion[1][i]
    for i in range(tam):
        poblacion[2][i]=poblacion[1][i]/sumatoria
    return sumatoria

def BubbleSort(poblacion,tam):
    j=0
    c=True
    while c==True:
        c=False
        for i in range(tam-1-j):
            if poblacion[1][i]<poblacion[1][i+1]:
                poblacion[0][i],poblacion[0][i+1]=poblacion[0][i+1],poblacion[0][i]
                poblacion[1][i],poblacion[1][i+1]=poblacion[1][i+1],poblacion[1][i]
                poblacion[2][i],poblacion[2][i+1]=poblacion[2][i+1],poblacion[2][i]
                c=True
        j+=1
        
def Torneo(tam):
    G1=[]
    G2=[]
    i=0
    while (i<3):
        a=random.randint(0,tam-1)
        b=random.randint(0,tam-1)
        if a in G1==True or a==b or b in G1==True or a in G2==True:
            continue
        G1.append(a)
        G2.append(b)
        i+=1
    G1.sort()
    G2.sort()
    return G1[0],G2[0]
    
def CruzaBinarioS(poblacion, padres,hijos):
    C1=poblacion[0][padres[0]]
    C2=poblacion[0][padres[1]]
    ch1,ch2=[],[]
    print("\n\t\tCruzamiento Binario Simulado")
    u=random.random()
    nx=60
    if (u<0.5): B=(2*u)**(1/(nx+1))
    else: B=(1/(2*(1-u)))**(1/(nx+1))
    for i in range(len(C1)):
        CH1=0.5*((1+B)*C1[i]+(1-B)*C2[i])
        CH2=0.5*((1-B)*C1[i]+(1+B)*C2[i])
        ch1.append(CH1)
        ch2.append(CH2)
        print(f"Gen {i+1}")
        print(f"Padre 1: {C1[i]}")
        print(f"Padre 2: {C2[i]}")
        print(f"Hijo 1: {CH1}")
        print(f"Hijo 2: {CH2}\n")
    hijos.append(ch1)
    hijos.append(ch2)
    
def CruzaAritmetica(poblacion, padres,hijos):
    C1=poblacion[0][padres[0]]
    C2=poblacion[0][padres[1]]
    ch1,ch2=[],[]
    B=random.random()
    print("\n\t\tCruzamiento Aritmético")
    for i in range(len(C1)):
        CH1=B*C1[i]+(1-B)*C2[i]
        CH2=B*C2[i]+(1-B)*C1[i]
        ch1.append(CH1)
        ch2.append(CH2)
        print(f"Gen {i+1}")
        print(f"Padre 1: {C1[i]}")
        print(f"Padre 2: {C2[i]}")
        print(f"Hijo 1: {CH1}")
        print(f"Hijo 2: {CH2}\n")
    hijos.append(ch1)
    hijos.append(ch2)
        
def CruzaMezcla(poblacion, padres,hijos):
    C1=poblacion[0][padres[0]]
    C2=poblacion[0][padres[1]]
    ch1,ch2=[],[]
    print("\n\t\tCruzamiento por Mezcla")
    for i in range(len(C1)):
        if (C1[i]>C2[i]):
            hmin=C2[i]
            hmax=C1[i]
        else:
            hmin=C1[i]
            hmax=C2[i]
        I=hmax-hmin
        a=random.random()
        CH1=random.uniform((hmin-I*a),(hmax+I*a))
        CH2=random.uniform((hmin-I*a),(hmax+I*a))
        ch1.append(CH1)
        ch2.append(CH2)
        print(f"Gen {i+1}")
        print(f"Padre 1: {hmax}")
        print(f"Padre 2: {hmin}")
        print(f"Hijo 1: {CH1}")
        print(f"Hijo 2: {CH2}\n")
    hijos.append(ch1)
    hijos.append(ch2)
    
def CruzaPlano(poblacion, padres,hijos):
    C1=poblacion[0][padres[0]]
    C2=poblacion[0][padres[1]]
    ch1,ch2=[],[]
    print("\n\t\tCruzamiento Plano")
    for i in range(len(C1)):
        if (C1[i]>C2[i]):
            hmin=C2[i]
            hmax=C1[i]
        else:
            hmin=C1[i]
            hmax=C2[i]
        CH1=random.uniform(hmin,hmax)
        CH2=random.uniform(hmin,hmax)
        ch1.append(CH1)
        ch2.append(CH2)
        print(f"Gen {i+1}")
        print(f"Padre 1: {C1[i]}")
        print(f"Padre 2: {C2[i]}")
        print(f"Hijo 1: {CH1}")
        print(f"Hijo 2: {CH2}\n")
    hijos.append(ch1)
    hijos.append(ch2)
    
        
def CruceLineal(poblacion,padres,hijos):
    C1=poblacion[0][padres[0]]
    C2=poblacion[0][padres[1]]
    Ch1,Ch2,Ch3=[],[],[]
    print("\n\t\tCruzamiento Lineal")
    for i in range(len(C1)):
        CH1=0.5*C1[i]+0.5*C2[i]
        CH2=(3/2)*C1[i]-(1/2)*C2[i]
        CH3=(1/2)*C1[i]+(3/2)*C2[i]
        Ch1.append(CH1)
        Ch2.append(CH2)
        Ch3.append(CH3)
        print(f"Gen {i+1}")
        print(f"Padre 1: {C1[i]}")
        print(f"Padre 2: {C2[i]}")
        print(f"Hijo 1: {CH1}")
        print(f"Hijo 2: {CH2}")
        print(f"Hijo 3: {CH3}\n")
    hijos.append(Ch1)
    hijos.append(Ch2)
    hijos.append(Ch3)
    
def ImprimirPoblacion(poblacion,tam):
    print("[X]\t\t\tF(X)\t\t\tFitness")
    for i in range(tam):
        print(f"\t\t{poblacion[1][i]}\t{poblacion[2][i]}")
        
def MutacionAzar(poblacion,tam,mutados):
    SeleccionCromosoma=random.randint(0,tam-1)
    CromosomaMutar=poblacion[0][SeleccionCromosoma]
    for i in range(len(CromosomaMutar)):
        CromosomaMutar[i]=random.uniform(-500,500)
    mutados.append(CromosomaMutar)

def MutacionNoUniforme(poblacion,tam,mutados,GActual,Gmax):
    SeleccionCromosoma=random.randint(0,tam-1)
    CromosomaMutar=poblacion[0][SeleccionCromosoma]
    print(f"\nCromosoma sin mutar({SeleccionCromosoma+1}): {CromosomaMutar}")
    for i in range(len(CromosomaMutar)):
        x=MaxMinGen(poblacion[0],i)
        fg=FG(GActual,Gmax)
        r=random.random()
        if (r<0.5):g1=(CromosomaMutar[i]+((x[0]-x[1])*fg))
        else: g1=(CromosomaMutar[i]-((x[0]+x[1])*fg))
        CromosomaMutar[i]=g1
    mutados.append(CromosomaMutar)
    print(f"\nCromosoma mutado({SeleccionCromosoma+1}): {CromosomaMutar}")

def FG(G,Gmax):
    b=5
    fg=(random.random()*(1-G/Gmax))**b
    return fg
    
def MaxMinGen(cromosomas,gen):
    genes=[]
    for i in range(len(cromosomas)):
        c=cromosomas[i]
        for i in range(len(c)):
            if (i==gen):
                genes.append(c[i])
    return max(genes),min(genes)
    
    
#def MutacionGaussiana():
        
    
def Generacion(generaciones,GeneracionActual,noGeneraciones):
    TamPob=100
    hijos=[]
    poblacion=Poblacion(TamPob)
    BubbleSort(poblacion,TamPob)
    calcularAptitud(poblacion,TamPob)
    ImprimirPoblacion(poblacion,TamPob)
    padres=Torneo(TamPob)
    CruzaBinarioS(poblacion,padres,hijos)
    CruzaAritmetica(poblacion,padres,hijos)
    CruzaMezcla(poblacion,padres,hijos)
    CruzaPlano(poblacion,padres,hijos)
    CruceLineal(poblacion,padres,hijos)
    NuevaGeneracion=Evaluacion(hijos,len(hijos))
    ImprimirPoblacion(NuevaGeneracion,len(hijos))
    mutacion=[]
    MutacionAzar(poblacion,TamPob,mutacion)
    MutacionNoUniforme(poblacion,TamPob,mutacion,GeneracionActual,noGeneraciones)
    NGMutados=Evaluacion(mutacion,len(mutacion))
    generaciones.append(poblacion)

def Main():
    print("\t\t\t T O R N E O \n")
    NoGeneraciones=50
    generaciones=[]
    generacionActual=1
    while generacionActual<=NoGeneraciones:
            Generacion(generaciones,generacionActual,NoGeneraciones)
            generacionActual+=1
    
    
Main()
```
